```python
# Prompt 1: Create a sample cybersecurity dataset with features like 'IP', 'Port', 'Protocol', 'Packet_Size', 'Label'
import pandas as pd

data = {
    "IP": ["192.168.1.1", "10.0.0.5", "172.16.0.8", "192.168.1.10", "10.0.0.12"],
    "Port": [80, 22, 443, 8080, 21],
    "Protocol": ["HTTP", "SSH", "HTTPS", "HTTP", "FTP"],
    "Packet_Size": [500, 120, 2000, 300, 50],
    "Label": ["Normal", "Attack", "Normal", "Attack", "Attack"]
}
df = pd.DataFrame(data)
df
```

```python
# Prompt 2: Check distribution of attack vs normal traffic
df["Label"].value_counts()
```

```python
# Prompt 3: Encode categorical features (Protocol, Label) for ML
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df["Protocol"] = encoder.fit_transform(df["Protocol"])
df["Label"] = encoder.fit_transform(df["Label"])
df
```

```python
# Prompt 4 (Fixed): Plot correlation heatmap of only numeric features
import seaborn as sns
import matplotlib.pyplot as plt

numeric_df = df.select_dtypes(include=["int64", "float64"])  # only numeric columns
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
# Prompt 5: Normalize packet size for better ML training
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df["Packet_Size"] = scaler.fit_transform(df[["Packet_Size"]])
df
```

```python
# Prompt 6 (Fixed): Split dataset into training and testing sets (dropping IP)
from sklearn.model_selection import train_test_split

X = df.drop(["Label", "IP"], axis=1)  # drop IP column
y = df["Label"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
# Prompt 7 (Fixed): Train a Decision Tree model to classify threats
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

print("Decision Tree Accuracy:", model.score(X_test, y_test))
```

```python
# Prompt 8: Train a Random Forest model for better performance
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
print("Random Forest Accuracy:", rf_model.score(X_test, y_test))
```

```python
# Prompt 9: Train a Logistic Regression model
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
print("Logistic Regression Accuracy:", lr_model.score(X_test, y_test))
```

```python
# Prompt 10: Evaluate model using confusion matrix
from sklearn.metrics import confusion_matrix
y_pred = rf_model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
```

```python
# Prompt 11: Build a simple Neural Network for threat detection
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

nn_model = Sequential([
    Dense(16, input_dim=X_train.shape[1], activation="relu"),
    Dense(8, activation="relu"),
    Dense(1, activation="sigmoid")
])

nn_model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
nn_model.fit(X_train, y_train, epochs=10, batch_size=2, verbose=1)
```

```python
# Prompt 12: Evaluate Neural Network model
loss, acc = nn_model.evaluate(X_test, y_test)
print("Neural Network Accuracy:", acc)
```

```python
# Prompt 13: Predict if a new incoming packet is malicious
sample = [[443, 1, 0.7]]  # (Port=443, Protocol=1, Packet_Size=0.7 normalized)
print("Prediction:", rf_model.predict(sample))
```

```python
# Prompt 14 (Fixed): Use Isolation Forest for anomaly detection
from sklearn.ensemble import IsolationForest

# Drop non-numeric column "IP"
features = df.drop(["Label", "IP"], axis=1)

iso = IsolationForest(contamination=0.2, random_state=42)
df["anomaly"] = iso.fit_predict(features)
df
```

```python
# Prompt 15: Visualize attack vs normal traffic
sns.countplot(x="Label", data=df)
plt.show()
```

```python
# Prompt 16: Visualize feature importance in Random Forest
import numpy as np
importances = rf_model.feature_importances_
features = X.columns
plt.bar(features, importances)
plt.title("Feature Importance")
plt.show()
```

```python
# Prompt 17: Apply KMeans clustering to group traffic patterns
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=42)
df["Cluster"] = kmeans.fit_predict(X)
df
```

```python
# Prompt 18: Use PCA for dimensionality reduction in cybersecurity data
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
reduced = pca.fit_transform(X)
plt.scatter(reduced[:,0], reduced[:,1], c=y)
plt.title("PCA of Cybersecurity Data")
plt.show()
```
